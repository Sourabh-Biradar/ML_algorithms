```python
# Logistic Regression

# ROC curve 
# ROC_AUC score

import warnings;warnings.filterwarnings('ignore')

from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000,random_state=11)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=11)
```

```python
# generating dummy model_prob output
dummy_model_prob = [0 for _ in range(len(y_test))]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model_prob = model.predict_proba(X_test)
```

```python
model_prob = model_prob[:,1]
```

```python
from sklearn.metrics import roc_auc_score,roc_curve

dummy_model_auc = roc_auc_score(y_test,dummy_model_prob)

actual_model_auc = roc_auc_score(y_test,model_prob)

print(dummy_model_auc)
print(actual_model_auc)
```

```python
# roc_curve

model_fpr,model_tpr,threshold = roc_curve(y_test,model_prob)
```

```python
# plot
import matplotlib.pyplot as plt
plt.plot(model_fpr,model_tpr,'r.')
plt.show()
```
